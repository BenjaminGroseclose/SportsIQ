```python
mlbPlayers = mlbPlayers.drop(mlbPlayers[mlbPlayers['Rk'] == 'Rk'].index)

print(mlbPlayers.shape)
```

```python
mlbPlayers['Player'] = mlbPlayers['Player'].str.replace('*', '')
mlbPlayers['Player'] = mlbPlayers['Player'].str.replace('#', '')
```
