```python
df = df.dropna(subset=['PositionCategory'])
df = df.dropna(subset=['Position'])
print(df.shape)

df['Team'] = df['Team'].replace(np.nan, None)
df['Jersey'] = df['Jersey'].replace(np.nan, None)
df['BatHand'] = df['BatHand'].replace(np.nan, None)
df['ThrowHand'] = df['ThrowHand'].replace(np.nan, None)
```

```python
"""
Index(['PlayerID', 'Status', 'TeamID', 'Team', 'Jersey', 'PositionCategory',
       'Position', 'FirstName', 'LastName', 'BatHand', 'ThrowHand', 'Height',
       'Weight', 'BirthDate', 'BirthCity', 'BirthState', 'BirthCountry',
       'PhotoUrl', 'InjuryStatus', 'FanDuelPlayerID', 'DraftKingsPlayerID',
       'FanDuelName', 'DraftKingsName'],
      dtype='object')
"""

for index, row in df.iterrows():
		date = None
		teamID = None

		if (row.Team is not None):
			cursor.execute("SELECT TeamID FROM MLB.Teams WHERE Abbreviation = ?;", row.Team)
			team = cursor.fetchone()
			teamID = team[0]

		if (row.BirthDate is not None and type(row.BirthDate) is str):
			date = datetime.strptime(row.BirthDate, "%m/%d/%Y %I:%M:%S %p")

		print(row.PlayerID)

		cursor.execute("""
								INSERT INTO MLB.Players
								(FirstName, LastName, Jersey, DateOfBirth, [PositionCategory], [Position], [Status], ExternalPlayerID, BatHand, ThrowHand, TeamID)
								VALUES(?,?,?,?,?,?,?,?,?,?,?);
								 """, 
								 row.FirstName, 
								 row.LastName, 
								 row.Jersey, 
								 date, 
								 row.PositionCategory, 
								 row.Position, 
								 row.Status, 
								 row.PlayerID, 
								 row.BatHand,
								 row.ThrowHand,
								 teamID)


cnxn.commit()
cursor.close()
```
