```python
# features = ['season', 'week', 'home_elo', 'home_fpi', 'home_rating',
#        'home_offense', 'home_defense', 'home_year_y',
#        'home_epa', 'home_epaAllowed', 'home_successRate',
#        'home_successRateAllowed', 'home_explosiveness',
#        'home_explosivenessAllowed', 'home_rank', 'home_points',
#       'away_elo', 'away_fpi', 'away_rating', 'away_offense',
#        'away_defense', 'away_epa',
#        'away_epaAllowed', 'away_successRate', 'away_successRateAllowed',
#        'away_explosiveness', 'away_explosivenessAllowed',
#        'away_rank', 'away_points']

# features = ['season', 'week', 'home_elo', 'home_fpi', 'home_rating',
#        'home_offense', 'home_defense', 'home_year_y',
#        'home_epa', 'home_epaAllowed', 'home_successRate',
#        'home_explosiveness',
#        'home_rank', 'home_points',
#       'away_elo', 'away_fpi', 'away_rating', 'away_offense',
#        'away_defense', 'away_epa',
#        'away_epaAllowed', 'away_successRate',
#        'away_explosiveness', 
#        'away_rank', 'away_points']

features = ['home_elo', 'home_fpi', 'home_rating',
        'home_defense', 'home_epa', 'home_successRate',
       'home_explosiveness',
       'home_rank',
      'away_elo', 'away_fpi', 'away_rating',
       'away_defense', 'away_epa', 'away_successRate',
       'away_explosiveness', 
       'away_rank']

# features = ['home_elo', 'home_fpi',
#          'home_epa', 
#       'away_elo', 'away_fpi',
#        'away_epa', ]


X = merged_df[features]
y = merged_df[['homePoints', 'awayPoints']]
```

```python
# Check for NaN values in X
print("NaN counts per column:")
print(X.isnull().sum()[X.isnull().sum() > 0])

# Option 1: Drop rows with NaNs
X = X.dropna()
y = y.loc[X.index]

# Option 2: Fill NaNs with column mean (or another value)
# X = X.fillna(X.mean())
# y = y.loc[X.index]
```

```python
# Train on all years except 2025, test on 2025
train_mask = merged_df['season'] != 2025
test_mask = merged_df['season'] == 2025

X_train = X[train_mask]
y_train = y[train_mask]
X_test = X[test_mask]
y_test = y[test_mask]
```

```python
type(X_train)
```

```python
model =  RandomForestRegressor(max_depth=8)
model.fit(X_train, y_train)


y_pred_train = model.predict(X_train)
train_mse = mean_squared_error(y_train, y_pred_train)
train_r2 = r2_score(y_train, y_pred_train)
print(f"Train MSE: {train_mse:.2f}, R2: {train_r2:.2f}")

y_pred_test = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred_test)
test_r2 = r2_score(y_test, y_pred_test)
print(f"Test MSE: {test_mse:.2f}, R2: {test_r2:.2f}")
```

```python
# Train a model for homePoints
gbr_home = Ridge()
gbr_home.fit(X_train, y_train['homePoints'])
home_train_pred = gbr_home.predict(X_train)
home_test_pred = gbr_home.predict(X_test)

# Train a model for awayPoints
gbr_away = Ridge()
gbr_away.fit(X_train, y_train['awayPoints'])
away_train_pred = gbr_away.predict(X_train)
away_test_pred = gbr_away.predict(X_test)

# Evaluate
from sklearn.metrics import mean_squared_error, r2_score
print("HomePoints - Train MSE:", mean_squared_error(y_train['homePoints'], home_train_pred), "R2:", r2_score(y_train['homePoints'], home_train_pred))
print("HomePoints - Test MSE:", mean_squared_error(y_test['homePoints'], home_test_pred), "R2:", r2_score(y_test['homePoints'], home_test_pred))
print("AwayPoints - Train MSE:", mean_squared_error(y_train['awayPoints'], away_train_pred), "R2:", r2_score(y_train['awayPoints'], away_train_pred))
print("AwayPoints - Test MSE:", mean_squared_error(y_test['awayPoints'], away_test_pred), "R2:", r2_score(y_test['awayPoints'], away_test_pred))
```
